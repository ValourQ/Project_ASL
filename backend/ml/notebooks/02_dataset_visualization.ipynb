# Visualizing a Landmark Frame

```python
import numpy as np
from pathlib import Path

DATASET = Path("../../../datasets/mutemotion")

landmarks = np.load(
    DATASET / "landmarks_V3.npz",
    allow_pickle=True
)

first_key = landmarks.files[0]
sample = landmarks[first_key]

print(first_key)
print(sample.shape)
```

```python
frame = sample[0]

print(frame.shape)
print(frame[:10])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))

plt.scatter(
    frame[:, 0],      # x
    -frame[:, 1],     # y (flip so it looks natural)
    s=4
)

plt.title("All 553 Landmarks")
plt.axis("equal")
plt.grid(True)

plt.show()
```

```python
left_hand = frame[0:21]
right_hand = frame[21:42]
pose = frame[42:75]
face = frame[75:]

print("Left hand:", left_hand.shape)
print("Right hand:", right_hand.shape)
print("Pose:", pose.shape)
print("Face:", face.shape)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,10))

plt.scatter(face[:,0], -face[:,1], s=2, label="Face")
plt.scatter(pose[:,0], -pose[:,1], s=30, label="Pose")
plt.scatter(left_hand[:,0], -left_hand[:,1], s=40, label="Left Hand")
plt.scatter(right_hand[:,0], -right_hand[:,1], s=40, label="Right Hand")

plt.legend()
plt.axis("equal")
plt.grid(True)
plt.title("Separated MediaPipe Landmarks")

plt.show()
```
