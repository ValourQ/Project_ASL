# Landmark Renderer

## Objective

In this notebook, we begin building the rendering engine for SignSync's **Text → Sign** module.

Instead of displaying prerecorded videos, SignSync will visualize MediaPipe landmark sequences as animated holographic skeletons.

This renderer will become the foundation of the Motion Mode in the Text → Sign translator.

### Goals

- Load one landmark sequence
- Render a single frame
- Connect landmarks into a skeleton
- Animate the sequence
- Prepare the renderer for future React integration

```python
import numpy as np
import matplotlib.pyplot as plt
import json

from pathlib import Path
```

```python
DATASET = Path("../../../datasets/mutemotion")

landmarks = np.load(
    DATASET / "landmarks_V3.npz",
    allow_pickle=True
)

with open(DATASET / "WLASL_parsed_data.json") as f:
    metadata = json.load(f)

print("Landmark Samples :", len(landmarks.files))
print("Metadata Samples :", len(metadata))
```

```python
sample_id = landmarks.files[0]

animation = landmarks[sample_id]

print("Sample ID :", sample_id)
print("Animation Shape :", animation.shape)
```
